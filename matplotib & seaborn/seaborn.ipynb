SEABORN


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = sns.load_dataset("tips")
```

```python
data
```

```python
sns.scatterplot(x= 'total_bill', y = 'tip', data = data)
```

```python
sns.relplot(x ='total_bill' , y ='tip', data = data)
```

```python
sns.relplot(x ='total_bill' , y ='tip', hue= 'time',data = data)
```

```python
sns.relplot(x ='total_bill' , y ='tip',hue= 'time',data = data)
```

```python
sns.relplot(x ='total_bill' , y ='tip', hue= 'time' ,style = 'smoker',kind = 'scatter',data = data)
```

```python
sns.relplot(x ='total_bill' , y ='tip', hue= 'time' ,style = 'smoker',kind = 'scatter',col ="smoker",data = data)
```

```python
sns.relplot(x ='total_bill' , y ='tip', hue= 'time' ,style = 'smoker',kind = 'scatter',col ="day",data = data)
```

```python
sns.relplot(y ='total_bill', x= 'tip', data = data)
plt.title("this is relplot")
plt.xlabel("Total_Bill", fontsize= 20, color ="black")
plt.ylabel("Tip", fontsize = 20, color = "pink")
plt.show()
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind= 'swarm', data= data)
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind= 'bo', data= data)
```

COUNT PLOT

```python
plt.figure(figsize=(5,5))
sns.countplot(x= 'day', data= data)
```

BOX PLOT

```python
sns.boxplot(x= 'day', y= 'total_bill', data =data)
```
